```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.scatterplot(x = df['x1'], y =df['x2'])
```

```python
from sklearn.cluster import KMeans
```

```python
list_inertias = []
for i in range(1,10):
    km = KMeans(n_clusters=i)
    km.fit(df)
    print(f"for i == inertia is {km.inertia_}")
    list_inertias.append(km.inertia_)
```

```python
sns.scatterplot(x=np.linspace(0,10,10),y=list_inertias)
```

```python
km = KMeans(n_clusters=3)
km.fit_predict(df)
km.inertia_
```

```python
x= df.iloc[:,:].values
```

```python
km.fit(x)
```

```python
predicted_y = km.predict(x)
```

```python
x[predicted_y == 0]
```

```python
len(predicted_y)
```

```python
sns.scatterplot(x=df['x1'],y = df['x2'],hue=predicted_y,color='red',palette="viridis")
plt.legend()
plt.title("K means clustering")
```
