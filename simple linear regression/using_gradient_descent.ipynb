```python
def gd_regressor(X_train, y_train, X_test, y_test, learning_rate=0.01, epochs=100000):
    m = 10  # random value)
    b = 10 #random value
    
    for i in range(epochs):
        y_pred = m * X_train + b  # Predicted values
        #derivative of loss function
        loss_slope_b = -2 * np.sum(y_train - y_pred) / len(X_train)
        loss_slope_m = -2 * np.sum((y_train - y_pred) * X_train) / len(X_train)
        b -= (learning_rate * loss_slope_b)
        m -= (learning_rate * loss_slope_m)
        if i % 10000 == 0:
            print(f"Epoch {i}: m = {m}, b = {b}")
 
```

```python
gd_regressor(x_train,y_train,x_test,y_test)
```
