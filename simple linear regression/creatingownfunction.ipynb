```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('placement.csv')
```

```python
df.head()
```

```python
sns.regplot(data=df,x='cgpa',y='package')
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(df['cgpa'],df['package'],test_size=0.2,random_state=42)
```

```python
x_train
```

```python
def find_m_and_b_fit(x_train,y_train):

    x_train = np.array(x_train)
    y_train = np.array(y_train)

    mean_x = x_train.mean()
    mean_y = y_train.mean()

    m = np.sum((x_train - mean_x) * (y_train-mean_y))/np.sum((x_train - mean_x) ** 2)
    b = mean_y - m* mean_x
    print(f"m is {m} and b is {b}")
    return m,b
```

```python
m,b = find_m_and_b_fit(x_train,y_train)
```

```python
def predict_values(x):
    return m*x+b
```

```python
y_predicted = predict_values(x_test)
```
