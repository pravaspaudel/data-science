```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
normal_values = np.random.normal(20,40,1000)
std = np.std(normal_values)
std
normal_values[4]
```

```python
#using seaborn library
sns.kdeplot(normal_values)
```

```python
#without using seaborn library
def density_plot(data_distribution):
    std = data_distribution.std()
    x = np.linspace(data_distribution.min(),data_distribution.max(),1000)
    coeffi = 1/(std*np.sqrt(2*np.pi))
    mean = data_distribution.mean()
    exponentpower = (-0.5)*(((x-mean)/std)**2)
    exponent_coeffi = coeffi * np.exp(exponentpower)
    plt.plot(x,exponent_coeffi)
```

```python
density_plot(normal_values)
```

```python
def check_using_QQ_plot(data_distribution):
    random_normal_values = np.random.normal(10,20,1000)
    x = np.sort(data_distribution)
    normal_quantiles = []
    data_qunatiles = []
    for i in range(1,101):
        n_value=np.percentile(random_normal_values,i)
        normal_quantiles.append(n_value)
        d_value = np.percentile(x,i)
        data_qunatiles.append(d_value)

    sns.regplot(x=normal_quantiles,y=data_qunatiles)
    
```

```python
#in iris dataset column

check_using_QQ_plot(iris_df['SepalWidthCm'])
```

```python
# for normal _distribution values the qq plot will be linear
check_using_QQ_plot(normal_values)
```

```python
def change_to_standard_normal_distribution(data):
    for i in range(len(data)):
        mean = data.mean()
        std = data.std()
        z = (data - mean)/std
    plt.title("standard normal distribution")
    plt.grid()
    sns.kdeplot(z)
```

```python
change_to_standard_normal_distribution(normal_values)
```

```python
change_to_standard_normal_distribution(iris_df['SepalWidthCm'])
```
